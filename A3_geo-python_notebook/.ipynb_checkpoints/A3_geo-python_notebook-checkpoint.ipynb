# LiDAR Point Cloud Processing with Python

## Overview

This notebook demonstrates basic LiDAR point cloud processing using Python libraries. We'll implement simple functions to:

1. **Read** LiDAR point cloud data
2. **Downsample** the point cloud to reduce data volume
3. **Classify** points as ground vs non-ground
4. **Generate a DEM** (Digital Elevation Model) from ground points

## Libraries Used

We'll primarily use:
- **laspy**: For reading LAS/LAZ files - [Documentation](https://laspy.readthedocs.io/en/latest/)
- **numpy**: For numerical operations - [Documentation](https://numpy.org/doc/stable/)
- **scipy**: For spatial operations and interpolation - [Documentation](https://docs.scipy.org/doc/scipy/)
- **matplotlib**: For visualization - [Documentation](https://matplotlib.org/stable/contents.html)
- **rasterio**: For raster operations and saving DEMs - [Documentation](https://rasterio.readthedocs.io/en/latest/)
- **something with gdal**

## Why These Libraries?

For geospatial projects involving LiDAR data, these libraries provide essential functionality:
- **laspy** is the standard for reading LiDAR files in Python
- **scipy.spatial** has efficient spatial data structures and algorithms
- **rasterio** has raster capabilities for saving DEMs


```python
import numpy as np
import laspy
from scipy.spatial import cKDTree
from scipy.interpolate import griddata
import matplotlib.pyplot as plt
import rasterio
from rasterio.transform import from_bounds
```

## Function 1: Read LiDAR Point Cloud

This function reads a LAS/LAZ file and extracts XYZ coordinates.

**Documentation reference**: [laspy.read()](https://laspy.readthedocs.io/en/latest/api.html#laspy.read)

```python
def read_lidar_data(file_path):
    las_data = laspy.read(file_path)
    
    points = np.vstack([las_data.x, las_data.y, las_data.z]).T
    print(f"Successfully read {len(points)} points from {file_path}")  
    print("Las data is the following:", las_data)
    
    return points
```

Here are simple and small LiDAR point cloud dataset examples so we can test _read_lidar_data_ function.

**LiDAR examples**: 

```python
lidar1 = r'_data\test_area_1.las'
lidar2 = r'_data\test_area_2.las'
lidar3 = r'_data\test_area_3.las'
lidar4 = r'_data\test_area_4.las'
```

```python
points = read_lidar_data(lidar2)
```

## Function 2: Downsample Point Cloud

This function reduces point density in order to make the rest of the processing flow and visualization easier. Otherwise, even for small and light point clouds the computation would be heavy and would take a lot of time 

**Documentation reference**: [numpy.unique()](https://numpy.org/doc/stable/reference/generated/numpy.unique.html)

```python
def downsample(points, grid_size=1.0):
    grid_x = np.floor(points[:, 0] / grid_size).astype(int)
    grid_y = np.floor(points[:, 1] / grid_size).astype(int)
    
    grid_indices = np.column_stack([grid_x, grid_y])
    
    unique_bins, unique_indices = np.unique(grid_indices, axis=0, return_index=True)
    
    downsampled_points = points[unique_indices]
    
    print(f"Downsampled from {len(points)} to {len(downsampled_points)} points")
    print(f"Reduction factor: {len(points)/len(downsampled_points):.2f}")
    
    return downsampled_points
```

```python
downsample_points = downsample(points, grid_size=1.0)
```

## Function 3: Ground Point Classification

Simple ground classification using height-based filtering and local neighborhood analysis.

**Documentation reference**: [scipy.spatial.cKDTree](https://docs.scipy.org/doc/scipy/reference/generated/scipy.spatial.cKDTree.html)

```python
def classification(points, height_threshold=2.0, neighborhood_radius=5.0):
    tree = cKDTree(points[:, :2]) 
    
    ground_mask = np.zeros(len(points), dtype=bool)
    
    print("Classifying ground points...")
    
    for i in range(len(points)):
        neighbor_indices = tree.query_ball_point(points[i, :2], neighborhood_radius)
        
        if len(neighbor_indices) > 1:
            neighbor_elevations = points[neighbor_indices, 2]
            local_min = np.min(neighbor_elevations)
            
            if points[i, 2] - local_min <= height_threshold:
                ground_mask[i] = True
    
    ground_points = points[ground_mask]
    
    print(f"Classified {np.sum(ground_mask)} points as ground ({np.sum(ground_mask)/len(points)*100:.1f}%)")
    
    return ground_mask, ground_points
```

```python
classified_points = classification(downsample_points, height_threshold=2.0, neighborhood_radius=5.0)

ground_mask = classified_points[0]
ground_points = classified_points[1]
```

## Function 4: Generate DEM from Ground Points

Create a Digital Elevation Model (DEM) using interpolation of ground points.

**Documentation reference**: [scipy.interpolate.griddata](https://docs.scipy.org/doc/scipy/reference/generated/scipy.interpolate.griddata.html)

```python
def create_dem(ground_points, resolution=1.0, method='linear'):
    x_min, x_max = ground_points[:, 0].min(), ground_points[:, 0].max()
    y_min, y_max = ground_points[:, 1].min(), ground_points[:, 1].max()
    
    x_grid = np.arange(x_min, x_max + resolution, resolution)
    y_grid = np.arange(y_min, y_max + resolution, resolution)
    X, Y = np.meshgrid(x_grid, y_grid)

    print(f"Interpolating DEM using {method} method...")
    Z = griddata(
        points=ground_points[:, :2],  # X,Y coordinates
        values=ground_points[:, 2],   # Z values
        xi=(X, Y),
        method=method,
        fill_value=np.nan
    )
    
    metadata = {
        'width': len(x_grid),
        'height': len(y_grid),
        'bounds': (x_min, y_min, x_max, y_max),
        'resolution': resolution,
        'transform': from_bounds(x_min, y_min, x_max, y_max, len(x_grid), len(y_grid))
    }
    
    print(f"DEM created: {Z.shape[1]}x{Z.shape[0]} pixels at {resolution}m resolution")
    
    return Z, metadata
```

```python
dem = create_dem(ground_points, resolution=1.0, method='linear')

z = dem[0]
metadata = dem[1]
```

## Function 5: Save DEM to GeoTIFF

Save the generated DEM as a GeoTIFF file for use in GIS applications.

**Documentation reference**: [rasterio.open()](https://rasterio.readthedocs.io/en/latest/api/rasterio.html#rasterio.open)

```python
def save_dem_geotiff(dem, metadata, output_path):
    with rasterio.open(
        output_path,
        'w',
        driver='GTiff',
        height=dem.shape[0],
        width=dem.shape[1],
        count=1,
        dtype=dem.dtype,
        transform=metadata['transform'],
    ) as dst:
        dst.write(dem, 1)
    
    print(f"DEM saved to {output_path}")
```

```python
save_dem_geotiff(z, metadata, "_results\my_dem2.tif")
```

## Function 6: Visualization

Create visualizations of the LiDAR processing results using matplotlib for plotting and analysis.

**Documentation reference**: [matplotlib.pyplot.scatter()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html), [matplotlib.pyplot.imshow()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.imshow.html)

```python
def visualize_results(original_points, ground_points, dem=None):
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))
    
    axes[0,0].scatter(original_points[:, 0], original_points[:, 1], 
                     c=original_points[:, 2], cmap='terrain', s=0.1, alpha=0.6)
    axes[0,0].set_title('Original Point Cloud (Top View)')
    axes[0,0].set_xlabel('X')
    axes[0,0].set_ylabel('Y')
    axes[0,0].set_aspect('equal')
    
    if ground_points is not None:
        axes[0,1].scatter(ground_points[:, 0], ground_points[:, 1], 
                         c=ground_points[:, 2], cmap='terrain', s=0.5)
        axes[0,1].set_title('Classified Ground Points')
        axes[0,1].set_xlabel('X')
        axes[0,1].set_ylabel('Y')
        axes[0,1].set_aspect('equal')
    
    axes[1,0].hist(original_points[:, 2], bins=50, alpha=0.7, label='All points')
    if ground_points is not None:
        axes[1,0].hist(ground_points[:, 2], bins=50, alpha=0.7, label='Ground points')
    axes[1,0].set_title('Elevation Distribution')
    axes[1,0].set_xlabel('Elevation')
    axes[1,0].set_ylabel('Frequency')
    axes[1,0].legend()
    
    if dem is not None:
        im = axes[1,1].imshow(dem, cmap='terrain', aspect='equal')
        axes[1,1].set_title('Generated DEM')
        plt.colorbar(im, ax=axes[1,1], label='Elevation')
    
    plt.tight_layout()
    plt.show()
```

```python
visualize_results(points, ground_points, z)
```
